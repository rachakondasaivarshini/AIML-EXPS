```python
# Basic syntax
print("Hello, PyCharm!")  # Output a string

# Data types
x: int = 5
y: float = 3.14
name: str = "Alice"
flag: bool = True

print(x, y, name, flag)
```

```python

# If-else
if x > 0:
    print("x is positive")
else:
    print("x is zero or negative")

# For loop
for j in range(3):
    print("Loop:", j)

# While loop
count = 0
while count < 3:
    print("Count:", count)
    count += 1

```

```python
# Functions

def greet(person: str) -> str:
    """Return a greeting string for the given name."""
    return f"Hello, {person}!"

print(greet(name))
```

```python
# List
fruits = ["apple", "banana", "cherry"]
fruits.append("date")

# Tuple (immutable)
coordinates = (10, 20)

# Dictionary
person = {"name": "Bob", "age": 30}
person["age"] = 31

# Set
primes = {2, 3, 5, 7}
primes.add(11)

print(fruits, coordinates, person, primes)
```

```python
# Libraries & Modules

import math
from random import randint

print("Square root of 16 is", math.sqrt(16))
print("Random integer between 1 and 10:", randint(1, 10))
```
