```python
from collections import deque

# --- Stack Implementation ---
print("STACK IMPLEMENTATION")

stack = deque()

# Push elements onto the stack
stack.append('A')
stack.append('B')
stack.append('C')

print("Stack after pushing A, B, C:", list(stack))
# Pop elements from the stack (LIFO)
print("Popped from stack:", stack.pop())
print("Stack after popping:", list(stack))

# --- Queue Implementation ---
print("\nQUEUE IMPLEMENTATION")

queue = deque()

# Enqueue elements into the queue
queue.append('1')
queue.append('2')
queue.append('3')

print("Queue after enqueuing 1, 2, 3:", list(queue))

# Dequeue elements from the queue (FIFO)
print("Dequeued from queue:", queue.popleft())
print("Queue after dequeuing:", list(queue))
```

```python
# Implement a priority queue using heapq to manage student scores by priority

import heapq

# List to store the heap
priority_queue = []

# Function to add a student to the priority queue
def add_student(name, score):
    # Use -score to simulate a max-heap
    heapq.heappush(priority_queue, (-score, name))
    print(f"Added: {name} with score {score}")

# Function to get the student with highest priority (highest score)
def pop_top_student():
    if priority_queue:
        score, name = heapq.heappop(priority_queue)
        print(f"Selected for top priority: {name} with score {-score}")
    else:
        print("No students in the queue.")

# Adding students
add_student("Alice", 92)
add_student("Bob", 85)
add_student("Charlie", 95)
add_student("Diana", 88)

print("\n--- Processing Students by Priority ---")

pop_top_student()
pop_top_student()
pop_top_student()
pop_top_student()
```

```python
from datetime import datetime, timedelta
from collections import Counter

# --- Using datetime class ---
print("DATETIME USAGE")

# Get current date and time
now = datetime.now()
print("Current datetime:", now)

# Format the datetime
formatted = now.strftime("%Y-%m-%d %H:%M:%S")
print("Formatted datetime:", formatted)

# Create a specific date
future_date = datetime(2025, 12, 25, 10, 30)
print("Future date (Christmas 2025):", future_date)

# Add 10 days to current date
ten_days_later = now + timedelta(days=10)
print("10 days from now:", ten_days_later)

# --- Using Counter class ---
print("\nCOUNTER USAGE")

# Count characters in a string
text = "banana"
char_count = Counter(text)
print("Character count in 'banana':", char_count)

# Count words in a sentence
sentence = "apple banana apple mango banana apple"
word_list = sentence.split()
word_count = Counter(word_list)

print("Word count:", word_count)

# Most common element
print("Most common word:", word_count.most_common(1))
```

```python
import numpy as np

# Create NumPy arrays
arr1 = np.array([10, 20, 30, 40])
arr2 = np.array([1, 2, 3, 4])

print("Array 1:", arr1)
print("Array 2:", arr2)

# Element-wise Addition
add = arr1 + arr2
print("Addition:", add)

# Element-wise Subtraction
sub = arr1 - arr2
print("Subtraction:", sub)

# Element-wise Multiplication
mul = arr1 * arr2
print("Multiplication:", mul)

# Element-wise Division
div = arr1 / arr2
print("Division:", div)

# Element-wise Power
power = arr1 ** 2
print("Power of Array 1 (squared):", power)
```

```python
import matplotlib.pyplot as plt

# --- Sample data ---
x = [1, 2, 3, 4, 5]
y_line = [10, 20, 15, 25, 30]
y_bar = [5, 7, 3, 8, 6]

# --- Line Graph ---
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)  # 1 row, 2 columns, plot 1
plt.plot(x, y_line, marker='o', color='blue', label='Line Data')
plt.title('Simple Line Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.legend()

# --- Bar Chart ---
plt.subplot(1, 2, 2)  # 1 row, 2 columns, plot 2
plt.bar(x, y_bar, color='green', label='Bar Data')
plt.title('Simple Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()

# --- Show plots ---
plt.tight_layout()
plt.show()
```
