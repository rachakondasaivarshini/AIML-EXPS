```
import os
import sys
try:
    from constraint import Problem
except ImportError:
    print("Installing python-constraint library...")
    import subprocess
    subprocess.check_call([sys.executable, "-m", "pip", "install", "python-constraint"])
    from constraint import Problem
courses = {
    'Math101': ('T1', 40, []),
    'Physics101': ('T2', 30, ['Mon-9']),
    'Chem101': ('T1', 20, []),
    'Bio101': ('T3', 25, ['Tue-10']),
    'CS101': ('T4', 60, []),
    'Eng101': ('T2', 35, [])
}

time_slots = ['Mon-9', 'Mon-10', 'Tue-9', 'Tue-10']

rooms = {
    'R1': 50,
    'R2': 30,
    'R3': 70
}

problem = Problem()
for course in courses:
    domain = []
    for slot in time_slots:
        for room in rooms:
            # Capacity Check
            if courses[course][1] <= rooms[room]:
                # Preference Check
                if not courses[course][2] or slot in courses[course][2]:
                    domain.append((slot, room))

    problem.addVariable(course, domain)
def room_conflict(*args):
    seen = set()
    for val in args:
        if val in seen:
            return False
        seen.add(val)
    return True

problem.addConstraint(room_conflict, list(courses.keys()))
def check_teacher_conflict(val1, val2, c1, c2):
    t1 = courses[c1][0]
    t2 = courses[c2][0]
    if t1 == t2:
        return val1[0] != val2[0]
    return True
course_list = list(courses.keys())
for i in range(len(course_list)):
    for j in range(i + 1, len(course_list)):
        course1 = course_list[i]
        course2 = course_list[j]

        problem.addConstraint(
            lambda val1, val2, c1=course1, c2=course2: check_teacher_conflict(val1, val2, c1, c2),
            (course1, course2)
        )

solution = problem.getSolution()

if solution:
    print("\n" + "="*40)
    print("FEASIBLE TIMETABLE FOUND")
    print("="*40)

    for course, (time, room) in sorted(solution.items(), key=lambda x: x[1]):
        teacher = courses[course][0]
        students = courses[course][1]
        print(f"🔹 {course:<10} | Teacher: {teacher} | Students: {students:<2} | Slot: {time:<6} | Room: {room}")
    print("="*40 + "\n")
else:
    print("\n No feasible timetable matches all specified constraints.\n")
```
